## KRR

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
from sklearn.kernel_ridge import KernelRidge
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming x and y are your feature matrix and target variable, respectively
# Replace them with your actual data

# Initialize variables to store the best metrics and corresponding random state
best_random_state = None
min_mae = float('inf')
min_rmse = float('inf')
max_r2 = float('-inf')

# Loop through different random states
for random_state in range(1, 500):  # You can adjust the range as needed
    # Split the data into training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=random_state)

    # Initialize the Kernel Ridge Regression model
    krr = KernelRidge()

    # Train the model
    krr.fit(x_train, y_train)

    # Predictions
    krr_pred_test = krr.predict(x_test)
    krr_pred_train = krr.predict(x_train)

    # Calculate metrics for test set
    krr_rmse_test = sqrt(mean_squared_error(y_test, krr_pred_test))
    krr_mae_test = mean_absolute_error(y_test, krr_pred_test)
    krr_r2_test = r2_score(y_test, krr_pred_test)

    # Calculate metrics for train set
    krr_rmse_train = sqrt(mean_squared_error(y_train, krr_pred_train))
    krr_mae_train = mean_absolute_error(y_train, krr_pred_train)
    krr_r2_train = r2_score(y_train, krr_pred_train)

    # Update the best metrics and random state if needed
    if krr_mae_test < min_mae:
        min_mae = krr_mae_test
        min_rmse = krr_rmse_test
        max_r2 = krr_r2_test
        best_random_state = random_state

# Print the best metrics and corresponding random state for KRR
print("Best Random State for KRR:", best_random_state)
print("Min Test RMSE for KRR:", min_rmse)
print("Min Test MAE for KRR:", min_mae)
print("Max Test R2 for KRR:", max_r2)

# Print metrics for training set for KRR
print("Min Train RMSE for KRR:", krr_rmse_train)
print("Min Train MAE for KRR:", krr_mae_train)
print("Max Train R2 for KRR:", krr_r2_train)

# Train the KRR model with the best random state
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=best_random_state)
krr = KernelRidge()
krr.fit(x_train, y_train)
krr_pred_test = krr.predict(x_test)

# Plotting the graph for KRR
plt.scatter(y_test, krr_pred_test, color='r', s=70, label='Test set for KRR')
sns.regplot(x=y_train, y=krr.predict(x_train), color='g', scatter_kws={'s': 70},
            line_kws={'color': 'blue'}, label='Train set for KRR')

plt.xlabel('Predicted Kappa for KRR', fontsize=12, weight='bold')
plt.ylabel('Experimental Kappa', fontsize=12, weight='bold')
plt.gca().set_aspect('equal', adjustable='box')
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.legend(fontsize=16)
plt.show()
```
